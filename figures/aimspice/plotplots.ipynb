```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Get all image files, excluding non-PNG files and the Jupyter Notebook file
image_files = [f for f in os.listdir('.') if f.endswith('.png') and f != 'leakage_current.png']

# Function to display an image
def display_image(file_name):
    img = Image.open(file_name)
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.title(file_name)
    plt.show()
if 'leakage_current.png' in os.listdir('.'):
    display_image('leakage_current.png')

# Loop through each image file and display it
for file_name in sorted(image_files):
    display_image(file_name)

# Display the special case image (leakage_current.png) if it exists
```
